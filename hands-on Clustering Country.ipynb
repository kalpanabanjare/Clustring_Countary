# Elbow method (To find no. of cluster)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
ls = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df1_scaled)
    ls.append([k, kmeans.inertia_])

# Create DataFrame with two columns
er = pd.DataFrame(ls, columns=["No. of clusters", "Error"])
print(er)
```

```python
# Plot the elbow plot
plt.figure(figsize=(8, 6))
plt.plot(er["No. of clusters"], er["Error"], marker='o')
plt.title("Elbow Plot")
plt.xlabel("Number of Clusters")
plt.ylabel("Error (Inertia)")
plt.show()
```

```python
plt.plot(er['No. of clusters'], er['Error'])
plt.xlabel('No. of clusters')
plt.ylabel('Error')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df1_scaled)
```

```python
output=kmeans.predict(df1_scaled)
```

# Hierarical Clustring

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
temp=pd.DataFrame(df1_scaled)
```

```python
Z = linkage(temp.head(50),method='single')
plt.figure(figsize=(12,16))
plt.title('Dendograms')
dendrogram1 = dendrogram(Z)
plt.show()
```

```python
Z = linkage(temp.head(50),method='single')
plt.figure(figsize=(12,16))
plt.title('Dendograms')
d= .35
plt.axhline(y=d, c='k')
dendrogram1 = dendrogram(Z)
plt.show()
```

```python
Z = linkage(temp.head(50),method='complete')
plt.figure(figsize=(12,16))
plt.title('Dendograms')
dendrogram1 = dendrogram(Z)
plt.show()
```

```python
Z = linkage(temp.head(50),method='complete')
plt.figure(figsize=(12,16))
plt.title('Dendograms')
d=1
plt.axhline(y=d, c='k')
dendrogram1 = dendrogram(Z)
plt.show()
```

```python
Z = linkage(temp.head(50),method='average')
plt.figure(figsize=(12,16))
plt.title('Dendograms')
d=.7
plt.axhline(y=d, c='k')
dendrogram1 = dendrogram(Z)
plt.show()
```
